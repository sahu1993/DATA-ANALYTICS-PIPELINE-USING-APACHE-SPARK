```python
#Name - Shivam Sahu
#team memberâ€™s Name - N/A

from csv import DictReader
import csv
import urllib

import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords

import urllib2
from bs4 import BeautifulSoup

with open("sports_urls.csv") as f:
    urls = [row["SportsData.data.web_url"] for row in DictReader(f)]
    
#print file
i = 1;
for url in urls:
    #print url
    filename = "Sportsfile"+str(i);
    i = i+1;
    nyData = open(filename,"a")
    try: 
        page = urllib2.urlopen(url).read()
    except Exception:
        print " "
    soup = BeautifulSoup(page,'html.parser')
    ptags = soup.find_all("p",text=True)   
    for par in ptags:
        #print(type(par.get_text(strip=True)))
        nyData.write((par.get_text(strip=True)).encode('utf8'))
```
